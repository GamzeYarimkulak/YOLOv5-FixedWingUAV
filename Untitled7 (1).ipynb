```python
from yt_dlp import YoutubeDL

video_url = "https://youtube.com/shorts/cm0JZVvgVoA?si=JF9TFwKvWvbHA_6_"

# Video indirme işlemi
ydl_opts = {
    'format': 'mp4',
    'outtmpl': './test_video.mp4',  # Çıktı dosyasının adı
}
with YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])

print("Video başarıyla indirildi: test_video.mp4")
```
